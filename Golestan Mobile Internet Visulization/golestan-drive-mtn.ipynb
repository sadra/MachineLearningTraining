```python
import pandas as pd 
import folium
from folium import LinearColormap
from folium.plugins import HeatMap
import branca

for_map = pd.read_csv('vill.csv')

max_amount = float(for_map['MTN'].max())

hmap = folium.Map(location=[37.3000000, 54.9900000], zoom_start=9, )

hm_wide = HeatMap( list(zip(for_map.Lat.values, for_map.Long.values, for_map.MTN.values)),
                   min_opacity=0,
                   max_val=max_amount,
                   radius=10, blur=10, 
                   max_zoom=1, 
                  gradient={.5: 'red', .9: 'yellow', 1: 'lime'}
                 )

legend = branca.colormap.LinearColormap(['red','yellow','lime'], index=[0.5, 0.9, 1.], caption='Internet Acess (MTN)')
hmap.add_child(legend)

#======= Market location ========
# for lat,lon,code in zip(for_map['Lat'],for_map['Long'],for_map['Code']): 
#     # Marker() takes location coordinates  
#     # as a list as an argument 
#     folium.Marker(location=[lat,lon],popup = code, 
#                   icon= folium.Icon(color="blue", 
#                   icon_color='yellow',icon = 'cloud')).add_to(hmap) 

hmap.add_child(hm_wide)
```
