```python
import pandas as pd 
import folium
from folium import LinearColormap
from folium.plugins import HeatMap
import branca

for_map = pd.read_csv('vill.csv')

max_amount = float(for_map['MCI'].max())

hmap = folium.Map(location=[37.3000000, 54.9900000], zoom_start=9, )

hm_wide = HeatMap( list(zip(for_map.Lat.values, for_map.Long.values, for_map.MCI.values)),
                   min_opacity=0,
                   max_val=max_amount,
                   radius=10, blur=10, 
                   max_zoom=1, 
                  gradient={.5: 'red', .9: 'yellow', 1: 'lime'}
                 )

legend = branca.colormap.LinearColormap(['red','yellow','lime'], index=[0.5, 0.9, 1.], caption='Internet Acess (MCI)')
hmap.add_child(legend)

hmap.add_child(hm_wide)
```

```python
hmap.save('heatmap-mci.html')
```
