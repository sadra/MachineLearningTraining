# Example 3 - Divergance and Overflow

## $$h(x) = x^5 -2x^4 + 2$$

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# Make Data
x_3 = np.linspace(-2.5, 2.5, 1000)

def h(x):
    return x**5 - 2*x**4 + 2
    
def dh(x):
    return 5*x**4 - 8*x**3
```

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001, max_interation=300):
    
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_interation):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        step_size = abs(new_x - previous_x)
        if step_size < precision:
            break
    return new_x, x_list, slope_list
```

```python
local_min, list_x, deri_list = gradient_descent(dh, initial_guess=-0.2, max_interation=70)

# Derivative and Scatter plot side by side
plt.figure(figsize=[15, 5])

# 1st Chart: Cost function
plt.subplot(1, 2, 1)

plt.title("Cost function", fontsize=17)
plt.xlabel("x", fontsize=16)
plt.ylabel("h(x): co", fontsize=16)
plt.plot(x_3, h(x_3), color="blue", linewidth=3)
plt.xlim([-1.2, 2.5])
plt.ylim([-1, 4])

plt.scatter(list_x, h(np.array(list_x)), color='red', s=100, alpha=0.85)

# 2nd Chart: Derivative function
plt.subplot(1, 2, 2)

plt.title("Slope of the Cost function", fontsize=17)
plt.xlabel("x", fontsize=16)
plt.ylabel("dh(x)", fontsize=16)
plt.plot(x_3, dh(x_3), color="skyblue", linewidth=5, alpha=0.7)
plt.xlim([-1, 2])
plt.ylim([-4, 5])
plt.grid()

plt.scatter(list_x, deri_list, color='red', s=100, alpha=0.85)

plt.show()

print('Local min occurs at:', local_min)
print('Cost at this minimum is:', h(local_min))
print('Number of steps:', len(list_x))
```
